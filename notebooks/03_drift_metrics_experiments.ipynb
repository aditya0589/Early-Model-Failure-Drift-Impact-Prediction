```

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sys
import os

# Add src to path
sys.path.append(os.path.abspath('..'))

from src.utils.config import config
from src.utils.logger import setup_logger

%matplotlib inline
sns.set_style("whitegrid")
```

```

from src.features.drift_metrics import calculate_drift_metrics
from src.data.drift_simulation import DriftSimulator
from src.data.load_data import load_data

df = load_data()
sim = DriftSimulator(df)
ref_df = df.sample(1000)
```

```

intensities = [0.0, 0.2, 0.5, 0.8, 1.0]
psi_scores = []

for i in intensities:
    batch, _ = sim.simulate_batch(1000, drift_type="covariate_shift", drift_intensity=i)
    metrics = calculate_drift_metrics(ref_df.drop(columns=['churn']), batch.drop(columns=['churn']))
    psi_scores.append(metrics['mean_psi'])

plt.plot(intensities, psi_scores, marker='o')
plt.xlabel("Drift Intensity")
plt.ylabel("Mean PSI")
plt.title("PSI vs Drift Intensity")
plt.show()
```
