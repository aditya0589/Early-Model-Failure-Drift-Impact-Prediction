Analyze how changing thresholds affects failure rate.

```python

try:
    df = pd.read_csv("../data/processed/failure_training_data.csv")
    
    thresholds = np.linspace(0.1, 0.9, 10)
    failure_rates = []
    
    for t in thresholds:
        rate = (df['drift_risk_score'] > t).mean()
        failure_rates.append(rate)
        
    plt.plot(thresholds, failure_rates, marker='o')
    plt.xlabel("Drift Score Threshold")
    plt.ylabel("Failure Rate (Simulated)")
    plt.title("Sensitivity Analysis")
    plt.show()
except:
    print("Data not found.")
```
